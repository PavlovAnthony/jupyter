```python
from sympy import *
import numpy as np
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
```

```python
# 
g = sin(x)**2-cos(x)**2
plot(g)
g
```

```python
# 
pprint(solveset(g, x), use_unicode = True)
```

```python
#  0
pprint(solveset(sin(x)**2-cos(x)**2<0, x, domain=S.Reals))
```

```python
#  0
pprint(solveset(sin(x)**2-cos(x)**2>0, x, domain=S.Reals))
```

```python
#
pprint(solveset(diff(g), x))
```

```python
# 
pprint(solveset(diff(g)<0, x, domain=S.Reals))
```

```python
# 
pprint(solveset(diff(g)>0, x, domain=S.Reals))
```
