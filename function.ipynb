```python
from sympy import *
import numpy as np
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
```

```python
#график функции
g = sin(x)**2-cos(x)**2
plot(g)
g
```

```python
#решения функции
pprint(solveset(g, x), use_unicode = True)
```

```python
#Функция меньше 0
pprint(solveset(sin(x)**2-cos(x)**2<0, x, domain=S.Reals))
```

```python
#Функция больше 0
pprint(solveset(sin(x)**2-cos(x)**2>0, x, domain=S.Reals))
```

```python
#экстремумы
pprint(solveset(diff(g), x))
```

```python
#убывание функции
pprint(solveset(diff(g)<0, x, domain=S.Reals))
```

```python
#возрастание функции
pprint(solveset(diff(g)>0, x, domain=S.Reals))
```
